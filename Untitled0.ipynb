```python
#Personalizando a máscara

#Dentro da máscara temos: {0:10.2f}

# | 0 -> posição que a variável aparece dentro do .format().[Pode ser deixado em branco.]
# | : -> separador a partir deles começamos as formatações
# | 10 -> separamos 10 casas caracteres dentro do print para mostrarmos o número. 
# | . -> a partir do ponto, estamos formatando os valores após a vírgula decimal.
# | 2 -> quantidade de casas decimais
# | f -> f para flout ou d para inteiro. 

# Exemplo: 

# Plote as variáveis abaixo 

a = 245
b = 2.835895

print('| {} | e | {} |'.format(a,b))

print('| {0:10} | e | {1:10.3f} |'.format(a,b))
```

```python
# Tuplas

# São variáveis compostas imutáveis.

# Estrutura da Tupla:

# As variáveis comuns conseguem armazenar apenas um valor de cada vez.
# Já as tuplas conseguem armazenar diversos valores dentro de uma só variável.
# Elas são úteis para o programador não precisar declarar infinitas variáveis.
# Para declarar uma tupla, basta colocar todos os valores entre parênteses e separados por vírgula.

# tupla = ('Var1, 'Var2', 'Var3')

# Exemplo 1

# Defina uma tupla com 3 strings e use o print para ver como ela se apresenta.

# Definindo uma tupla:

t1 = ('Lanche', 'Batata', 'Refri')

# Imprimindo a tupla:

print(t1)
```

```python
# Exemplo 2

# Agora defina outra tupla com valores numéricos.

# Declarando tupla:
t2= (1, 3, 5, 7)

# Iprimindo tupla:
print(t2)
```

```python
# Exemplo 3
# Agora defina outra tupla com diversos tipos de dados inclusos.

# Defina a tupla:
t3=('Lanche', 10.9, 3, True)
#Imprimindo a tupla:
print(t3)
```

```python
# Acessando valores

# Para acessar os valores das tuplas, basta colocar o índice da tupla desejado entre colchetes.
# Cada valor dentro da tupla fica salvo como um índice que começa em 0 e vai aumentando de 1 em 1 até o último termo da tupla.

# Então para a tupla abaixo:

#     [0] [1] [2] [3]
# t2 =(1,  3,  5,  7)

# Os termos dela são:

#t2[0] == 1
#t2[1] == 3
#t2[2] == 5
#t2[3] == 7

# exemplo 1 
# Para a tupla acima. Tente imprimir cada um de seus valores.

# Imprimindo
print(t2[0])
print(t2[1])
print(t2[2])
print(t2[3])
```

```python
# calculo com elementos de uma tupla

print(t2[0] + t2[1])
```

```python
# Exemplo 2
# Imprima a soma de cada valor dessa tupla.

# Somando:

soma = t2[0] + t2[1] + t2[2] + t2[3]

# Impriindo:
print(soma)
```

```python
# Acessando múltiplos valores:
# Também é possível acessar mais de um valor ao mesmo tempo, colocando nos colchetes de onde até onde você quer acessar.

t2[0:2] # são os dois primeiros valores dessa tupla.
```

```python
t2[2:4] # são o terceiro e o quarto valores dessa tupla.
```

```python
t2[:3] # são os 3 primeiros valores dessa tupla.
```

```python
t2[2:] # é o terceiro e o último valor dessa tupla.
```

```python
t2[-1] # é o último valor dessa tupla.

# Lembre-se que como no range, o último valor dentro do colchete não é acessado.
```

```python
# Revisando os principais conceitos estatísticos 

# Variável aleatória(v.a.)

# Em um estudo, há sempre a necessidade de focar em um ou mais atributos(características)
# dos elementos que compõem minha população.

# Na situação problema em questão o que vem a ser a minha população?( R.: 30000 clientes)

# Atributos quantitativos: valor_credito
# atributos qualitativos: status_peg_Set_2005

# Definição de variável aleatória:

# É uma função que associa cada elemento de uma população a um número.

# Na prática o que é uma variável aleatória
credit.drop_duplicates(subset=None,keep='first',inplace=True) 
credit['Idade']
```

```python
# No exemplo acima a variável aleatória é definida por:

# x = {24,26,34,37,57,...,39,43,37,41,46}

# Caracterização de uma V.A.:
# Posição(tendência central)
# Dispersão
# Forma

# Vamos então entender o que são as medidas de tendência central

# Média - soma de todos os elementos divididos pela quantidade de elementos.

credit['valor_crédito'].mean()
```

```python
# Mediana - é o ponto que divide a distribuição em partes iguais.

credit['valor_crédito'].median()
```

```python
# Moda - valores que ocorrem com maior frequência.
credit['valor_crédito'].mode()
```

```python
# Vamos agora entender o que são as medidas de dispersão.

# Variância - mostra a distância dos dados em relação à média. 
# Quanto maior a variância, mais distante da média.
# Quanto menos a variãncia, mais próximos estão da média.

credit['valor_crédito'].var()
```

```python
# Desvio padrão - mostra a variabilidade dos dados em relação à média.
# Mostra quanto um conjunto de dados é uniforme.
# Quanto mais próximo de 0 o desvio padrão, mais homogêneos são os dados.

credit['valor_crédito'].std()
```

```python
# Gráfico com matplotlib - Plotagem de funções matemáticas(integração com Numpy)

# Criar uma matriz com a função linspace da numpy
t = np.linspace(0.0,2.0, 100)
t

# criar outras duas matrizes com diferentes funções trigonométricas

f1 = np.sin(np.pi * t)
f2 = np.sin(2 * np.pi * t)


plt.figure(figsize=(10,8))

plt.plot(t,f1, linewidth=5.0, label = 'função 1')
plt.plot(t,f2, linewidth=5.0, label = 'função 2')

plt.legend()

plt.title('SENOIDES')

plt.grid(b=True, color='gray', linestyle= '--', linewidth=0.5);
```
